```python
import requests
import pandas as pd 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
url = "http://www.index.go.kr/potal/main/EachDtlPageDetail.do?idx_cd=2768"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
result = soup.find_all('tr')
result = [line.get_text().strip() for line in result]


result = result[3:18]

for i in range(0, len(result)):
    result[i]=result[i].split('\n')
for i in range(0, len(result)):
    del result[i][0]
for i in range(0, len(result)):
    for j in range(0, len(result[i])):
        result[i][j] = int(result[i][j].replace(',',''))
        
data = pd.DataFrame(
    data = result,
    index = ['계','지체장애','시각장애','청각, 언어','지적 장애','뇌병변장애','자폐성장애', '정신 장애','신장장애','심장장애','호흡기장애','간장애','안면장애','장애,요루장애','뇌전증장애'],
    dtype = float )

data.columns = ['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018']

data
```

```python
plt.rcParams["figure.figsize"] = (30,20)
disorder09 = data.groupby('2009').describe()
disorder10 = data.groupby('2010').describe()
disorder11 = data.groupby('2011').describe()
disorder12 = data.groupby('2012').describe()
disorder13 = data.groupby('2013').describe()
disorder14 = data.groupby('2014').describe()
disorder15 = data.groupby('2015').describe()
disorder16 = data.groupby('2016').describe()
disorder17 = data.groupby('2017').describe()
disorder18 = data.groupby('2018').describe()
fig = plt.figure()
axes1 = fig.add_subplot(5,2,1)
axes2 = fig.add_subplot(5,2,2)
axes3 = fig.add_subplot(5,2,3)
axes4 = fig.add_subplot(5,2,4)
axes5 = fig.add_subplot(5,2,5)
axes6 = fig.add_subplot(5,2,6)
axes7 = fig.add_subplot(5,2,7)
axes8 = fig.add_subplot(5,2,8)
axes9 = fig.add_subplot(5,2,9)
axes10 = fig.add_subplot(5,2,10)
axes1.set_title('2009 average of disorder')
axes2.set_title('2010 average of disorder')
axes3.set_title('2011 average of disorder')
axes4.set_title('2012 average of disorder')
axes5.set_title('2013 average of disorder')
axes6.set_title('2014 average of disorder')
axes7.set_title('2015 average of disorder')
axes8.set_title('2016 average of disorder')
axes9.set_title('2017 average of disorder')
axes10.set_title('2018 average of disorder')
axes1.plot(disorder09)
axes2.plot(disorder10)
axes3.plot(disorder11)
axes4.plot(disorder12)
axes5.plot(disorder13)
axes6.plot(disorder14)
axes7.plot(disorder15)
axes8.plot(disorder16)
axes9.plot(disorder17)
axes10.plot(disorder18)
fig.tight_layout()
```
